```python
import os, sys, subprocess
caminho_mod = subprocess.check_output([sys.executable, "-m", "amplpy.modules", "path"]).decode().strip()
os.environ["PATH"] += os.pathsep + caminho_mod
from amplpy import AMPL
ampl = AMPL()
ampl.eval("option solver cbc;")
print("AMPL configurado com solver CBC")
```

```python
ampl.eval("""
reset;
var x1 >= 0;
var x2 >= 0;
maximize Lucro: 20*x1 + 50*x2;
subject to MateriaPrima: 2*x1 + 4*x2 <= 240;
subject to LimiteVendasA: x1 <= 100;
subject to Proporcao: 0.2*x1 - 0.8*x2 >= 0;
""")
ampl.solve()
print(f"Lucro ótimo: {ampl.getObjective('Lucro').value()}")
print(f"x1 (Produto A): {ampl.getVariable('x1').value()}")
print(f"x2 (Produto B): {ampl.getVariable('x2').value()}")
```
