```python
import os, sys, subprocess
caminho_mod = subprocess.check_output([sys.executable, "-m", "amplpy.modules", "path"]).decode().strip()
os.environ["PATH"] += os.pathsep + caminho_mod
from amplpy import AMPL
ampl = AMPL()
ampl.eval("option solver cbc;")
print("AMPL configurado com solver CBC")
```

```python
ampl.eval("reset;")

# Definição das Variáveis (Toneladas de chapas e barras)
ampl.eval("""
var x1 >= 0;  # Chapas
var x2 >= 0;  # Barras

# Função Objetivo: Maximizar Lucro ($40 por chapa e $35 por barra)
maximize Z: 40*x1 + 35*x2;

# Restrições
# 1. Capacidade de Produção (800 chapas OU 600 barras)
subject to capacidade: (x1/800) + (x2/600) <= 1;

# 2. Demanda Máxima Diária
subject to demanda_chapas: x1 <= 550;
subject to demanda_barras: x2 <= 580;
""")

# Execução e Solver
ampl.solve()

# Exibição dos Resultados
print(f"--- Resultado Alumco ---")
print(f"Lucro total ótimo: ${ampl.getObjective('Z').value()}")
print(f"Produção de Chapas (x1): {ampl.getVariable('x1').value()} toneladas")
print(f"Produção de Barras (x2): {ampl.getVariable('x2').value()} toneladas")
```
